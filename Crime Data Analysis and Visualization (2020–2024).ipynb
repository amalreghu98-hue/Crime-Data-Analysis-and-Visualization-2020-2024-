```python
# check and handle null value in vict sex

rows_before = len(df)

df = df.dropna(subset=['Vict Sex'])

rows_after = len(df)

print("Rows removed:", rows_before - rows_after)
```

```python
df.isna().sum()
```

```python
# Fill missing Premis Cd with its mode
df['Premis Cd'] = df['Premis Cd'].fillna(df['Premis Cd'].mode()[0])

# Fill missing Premis Desc with its mode
df['Premis Desc'] = df['Premis Desc'].fillna(df['Premis Desc'].mode()[0])
```

```python
# check and handle null value in Weapon Used Cd
rows_before = len(df)

df = df.dropna(subset=['Weapon Used Cd'])

rows_after = len(df)

print("Rows removed:", rows_before - rows_after)
```

```python
# Fill missing text with a placeholder in cross street column
df['Cross Street'] = df['Cross Street'].fillna('UNKNOWN')
```

#### Detect and Handle Outliers Using IQR

```python
df.describe()
```

```python
Q1 = df['Vict Age'].quantile(0.25)
Q3 = df['Vict Age'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = df[(df['Vict Age'] < lower_bound) | (df['Vict Age'] > upper_bound)]

print("Number of outliers:", len(outliers))
outliers.head()
```

```python
print(outliers['Vict Age'])
```

```python
# For multiple numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"\n{col} - {len(outliers)} outliers")
    print(outliers[col].values)
```

```python
# Remove Outliers from All Numerical Columns
df_clean = df.copy()

for col in num_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df_clean = df_clean[
        (df_clean[col] >= lower) &
        (df_clean[col] <= upper)
    ]

print("Original Shape:", df.shape)
print("After Removing Outliers:", df_clean.shape)
```

#### 11. Which crime types remain unresolved in different areas?

```python
# BY USING HEATMAP
# 1. Filter for unresolved cases
unresolved = df[df["Status"] != "IC"]

# 2. Create the pivot table aggregating crime counts by Area and Description
pivot_table = pd.pivot_table(
    unresolved,
    values="Crm Cd",
    index="AREA NAME",
    columns="Crm Cd Desc",
    aggfunc="count",
    fill_value=0,
    observed=False,
)

# 3. Get the top 10 most frequent unresolved crimes to keep the heatmap clean
# (Fixed: Using 'unresolved' instead of 'df' to prevent potential KeyErrors)
top_crimes = unresolved["Crm Cd Desc"].value_counts().head(10).index

# 4. Filter pivot table columns safely using .loc
filtered_pivot = pivot_table.loc[:, top_crimes]

# 5. Plot the heatmap
plt.figure(figsize=(15, 8))
sns.heatmap(filtered_pivot, cmap="Reds", annot=True, fmt="d", cbar=True)

# 6. Formatting titles and adjusting rotated labels for readability
plt.title("Unresolved Crimes Across Areas (Top 10 Crimes)", fontsize=16, pad=15)
plt.xlabel("Crime Type", fontsize=12)
plt.ylabel("Area", fontsize=12)

# Rotate x-axis labels so they don't overlap
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
```

##### Insight

###### * Primary Unresolved Hotspot (77th Street Area): 77th Street shows the highest volume of unresolved violent and domestic incidents across nearly all top categories. This includes 2,592 cases of Intimate Partner - Simple Assault and 1,993 cases of Assault with Deadly Weapon.
###### * Intimate Partner - Simple Assault consistently registers as the most prominent unresolved crime category across almost all geographic areas, with notable concentrations in 77th Street (2,592), Olympic (1,946), and West Valley (1,717).
###### * Aside from 77th Street, areas like Central, Olympic, and Southwest stand out with uniformly high counts (darker red blocks) across Battery - Simple Assault and Assault with Deadly Weapon.
###### * Severe offenses like Rape, Forcible and Child Abuse (Physical) show significantly lower counts (lighter purple/pink shades, ranging mostly below 200 cases per area) compared to assault categories, though they remain present across every single jurisdiction.

##### Insight

###### * Perfect Structural Collinearity (AREA & Rpt Dist No = 1.00): There is a perfect linear correlation of 1.00 between the geographic Area code and the Reporting District Number. This indicates structural redundancy; the Rpt Dist No is directly derived from or contains the AREA identifier.
###### * Strong Categorical Relationship (Part 1-2 & Crm Cd = 0.73): A high positive correlation of 0.73 exists between the crime code (Crm Cd) and the crime classification level (Part 1-2).
###### * Geographic Baseline Redundancy (LAT & LON = -1.00): A perfect negative correlation of -1.00 appears between Latitude and Longitude. In real-world geography, LAPD coordinates do not form a perfectly straight diagonal line. This artifact typically happens because a large number of missing or unrecorded location entries are filled with placeholder 0.0 values for both latitude and longitude, skewing the correlation coefficient.
###### * Independence of Victim Characteristics: The correlation between Vict Age and all other operational or structural variables (like crime code, time of occurrence, or area) sits very close to 0.00. This implies that crime types or geographic locations are not linearly dependent on the victim's age alone.
