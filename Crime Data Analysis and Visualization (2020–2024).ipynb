#### Detect and Handle Outliers Using IQR

```python
df.describe()
```

```python
Q1 = df['Vict Age'].quantile(0.25)
Q3 = df['Vict Age'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = df[(df['Vict Age'] < lower_bound) | (df['Vict Age'] > upper_bound)]

print("Number of outliers:", len(outliers))
outliers.head()
```

```python
print(outliers['Vict Age'])
```

```python
# For multiple numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"\n{col} - {len(outliers)} outliers")
    print(outliers[col].values)
```

```python
# Remove Outliers from All Numerical Columns
df_clean = df.copy()

for col in num_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df_clean = df_clean[
        (df_clean[col] >= lower) &
        (df_clean[col] <= upper)
    ]

print("Original Shape:", df.shape)
print("After Removing Outliers:", df_clean.shape)
```

Because the vertical axis (Crime Count) peaks at exactly 1, the "peaks and valleys" do not represent a fluctuating volume of crimes. Instead, the chart is simply mapping a binary state: 1 means at least one crime occurred for that area/year combination, and 0 means none did.
